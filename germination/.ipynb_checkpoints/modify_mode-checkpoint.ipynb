```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import time
import datetime
%matplotlib inline
```

```python
# 調整高斯參數
kernel_size = (3, 3)  # 高斯模糊矩陣大小
sigma = 3 # 高斯模糊標準差參數(0=自動)
slice_img = cv2.imread("3G1_rect1.jpg")  # 整體成像預覽
img_size = (1100, 650)  # 顯示視窗大小

# slice_img = cv2.imread("./3G1/0-8.jpg")  # 個別成像預覽
# img_size = (100, 100)  # 顯示視窗大小

plt.xticks([]), plt.yticks([]) # to hide tick values on X and Y axis 
plt.imshow(slice_img) 
plt.show()

grayImage = cv2.cvtColor(slice_img, cv2.COLOR_BGR2GRAY)  # gray(0-255) 圖像轉換灰階
grayImage = cv2.GaussianBlur(grayImage, kernel_size, sigma)  # GaussianBlur  圖校進行模糊化(高斯)
(thresh, blackAndWhiteImage) = cv2.threshold(grayImage,  150, 255, cv2.THRESH_BINARY)  # black and white(0, 255)  圖像透過thershold閾值進行二元化 
slice_percent =  round(sum(sum(blackAndWhiteImage == 0))/100, 2)  # 計算黑色比例(作物占比)
print(slice_percent)
plt.xticks([]), plt.yticks([]) # to hide tick values on X and Y axis 
plt.imshow(blackAndWhiteImage) 
plt.show()


cv2.namedWindow("output", cv2.WINDOW_NORMAL)        # Create window with freedom of dimensions
imS = cv2.resize(blackAndWhiteImage, img_size)    # Resize image
cv2.imshow("output", imS)                           # Show image
cv2.waitKey(0)                                      # Display the image infinitely until any keypress

cv2.destroyAllWindows()
```

```python
# 照片裁切個別另存
img = cv2.imread("./3G1_rect1.jpg")
plt.imshow(img, cmap = 'gray', interpolation = 'bicubic') 
plt.xticks([]), plt.yticks([]) # to hide tick values on X and Y axis 
# plt.plot([200,300,400],[100,200,300],'c', linewidth=5) 
plt.show()
x = img.shape[0]
y = img.shape[1]
# 裁切區域的 x 與 y 座標（左上角）
top = 0  # 最上面(+h)
left = 0  # 最左邊(+w)
plus = 100
print(x, y)
for i in range(13):
    for j in range(22):
        slice_img = img[top:top+plus, left:left+plus]
        name = str(i)+'-'+str(j)+'.jpg'
        dirs = './3G1/'  # 紀德艦資料夾(懶的自動怎樣)
        cv2.imwrite(dirs+name, slice_img)
        left = left + plus
    top = top +plus
    left = 0
print('done')
```
